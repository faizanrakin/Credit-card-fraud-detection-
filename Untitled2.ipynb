```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('creditcard.csv')
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
# checking for any null values
df.isnull().sum().max()
```

```python
df.drop(columns = 'Time', inplace =True)
```

```python
Normal = int(df[df.Class==0]['Class'].value_counts())
Fraud = int(df[df.Class==1]['Class'].value_counts())
print('Normal : {:.2f}%'.format((Normal/(Normal+Fraud))*100))
print('Fraud : {:.2f}%'.format((Fraud/(Normal+Fraud)*100)))
```

```python
sns.set_theme(style='darkgrid')
sns.countplot(data=df,x='Class')
plt.title('Class distribution \n (0: No Fraud | 1: Fraud)',Fontsize = 14)
plt.show()
# Class imbalance needs to be taken care before applying model
```

```python
# Outlier detection

sns.scatterplot(data=df[df['Class']==0]['Amount'] ,label ='Not Fraud')
sns.scatterplot(data=df[df['Class']==1]['Amount'] ,label ='Fraud')
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
new_col = df['Amount'].to_numpy()
new_col = new_col.reshape(-1,1)
```

```python
scaler = StandardScaler()
scaler.fit(new_col)
new_col = scaler.transform(new_col)
```

```python
df['Amount'] = new_col
df.head(7)
```

```python
df.describe()
```

```python
x = df.drop(columns = 'Class')
y = df.Class
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.6, test_size=0.4, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=200, max_depth=15, random_state=0)
```

```python
model = clf.fit(x_train,y_train)
```

```python
y_predict = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,f1_score
```

```python
acc = accuracy_score(y_predict, y_test)
c_matrix = confusion_matrix(y_predict, y_test)
f1 = f1_score(y_predict, y_test)
result = pd.DataFrame([acc,f1], index = ['acc','f1'])
```
